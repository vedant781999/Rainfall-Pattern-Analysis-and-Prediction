```python
#Here we read the dataset containing the rainfall data
#of states all over India.
#Import the appropriate file for that region.
data = pd.read_csv("../Data/central-India_rainfall_act_dep_1901_2016.csv")
data
```

# **Data Cleaning**

## Missing data

```python
dataset = data.copy()
mean_rainfall = dataset.mean(axis=0,skipna=True)
std_rainfall = dataset.std(axis=0,skipna=True)
std_rainfall
```

```python
for row in range(len(dataset)):
  for col in dataset.columns:
    if pd.isna(dataset.loc[row,col]):
      dataset[row,col] = mean_rainfall[col]

dataset
```

# Data Preprocessing

## **Standardization**

```python
#Here we standardize the data grouped by states for all 12 months.
standardized_dataset=dataset.copy()
standardized_dataset=standardized_dataset.values
for row in range(0,len(standardized_dataset)):
  for col in range(1,len(standardized_dataset[1])):
    data_mean = mean_rainfall[col]
    data_std = std_rainfall[col]
    standardized_dataset[row,col]=(standardized_dataset[row,col]-data_mean)/data_std
```

```python
standardized_dataset=pd.DataFrame(standardized_dataset)
standardized_dataset
```
